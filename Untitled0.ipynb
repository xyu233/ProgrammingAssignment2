```python
import pandas as pd
import matplotlib.pyplot as plt
plt.close("all")
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("/content/companies.csv")
print(df.loc[df['country'].str.contains("USA")])
```

```python
temp = df.sort_values(by='annualRevenue', ascending=False).copy()
temp.plot.bar(x="companyName", y="annualRevenue")
plt.show()
```

```python
df['new column'] = df['city'] + ", " +df['state'] + ', ' +df['country']
print(df)
```

```python
df.plot.scatter(x="companyName", y="employees", s=df["employees"]*0.0005)
plt.xticks(rotation=45)
plt.show()
```
